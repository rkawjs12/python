```python
text1 = ' python programming '
#'g'를 출력하세요
print(text1[-2])
#print(text[text1.find('g')])

#'pro'를 출력하세요
print(text1[text1.find('pro'):text1.find('pro')+len('pro')])


#'gram'이; 포함되어 있는지 확인하세요.
print('gram' in text1)
#print(text1.find('gram'))


#모두 대문자로 바꾸세요
print(text1.upper())

#'p'자의 개수를 출력하세요
print(text1.count('p'))

#스페이스(공백)을 기준으로 분리하세요
print(text1.split())

#'-'를 기준으로 두 단어를 연결하세요
print('-'.join(text1.split()))

#양쪽의 공백을 제거하세요
print(text1.strip())
#'python'을 'hello'로 바꾸세요
print(text1.replace('python', 'hello'))
```

```python
solarsys = ['태양','수성','금성','지구','화성','목성','지구']
solarsys1 = ['토성','천왕성','해왕성']
planet = '지구'

#solarsys에서 ['수성','금성','지구']를 출력
print(solarsys[1:4])

#        에서 지구의 위치값 출력
print(solarsys.index('지구'))

#        에서 화성을 해왕성으로 수정
print(solarsys.index('화성'))
solarsys.remove('화성')
solarsys.insert(4,'해왕성')
print(solarsys)

#       에서 수성 자리에 명왕성을 삽입
print(solarsys.index('수성'))
solarsys.insert(1,'명왕성')
print(solarsys)

#        에 토성을 맨 끝에 첨가
solarsys.append('토성')
print(solarsys)

#        에 맨 끝 요소를 제거하고 리턴
print(solarsys.pop(len(solarsys)-1))

#        에서 1번째에서 3번째까지 삭제
del solarsys[1:4]
print(solarsys)

#        에서 인덱스 값 2번 요소를 삭제
print(solarsys.pop(2))

#        와     1을 합하시오
solarsys.extend(solarsys1)
print(solarsys)

#        를 오름차순으로 정렬하세요
solarsys.sort()
print(solarsys)

#        를 내림차순으로 정렬하세요
solarsys.sort(reverse=True)
print(solarsys)

#         원소의 순서를 바꾸세요
solarsys.reverse()
print(solarsys)

#        에 만리장성이 있는지 확인하세요
print('만리장성' in solarsys)

#        를 3번 반복하세요
print(solarsys*3)

#        의 길이를 리턴하세요
print(len(solarsys))

#        에서 지구의 갯수를 출력하세요.
print(solarsys.count('지구'))

#        의 모든 원소를 지우세요
solarsys.clear()
print(solarsys)

#         리스트를 삭제하세요
del solarsys
print(solarsys)
```

```python
#dictionary

#mydict이라는 빈 딕셔너리를 만드세요
mydict = {}
print(mydict)

#red:1, blue:3 greem:4 세 요소를 추가하세요
mydict = {'red':1, 'blue':3, 'green':4 }
print(mydict)

#green의 value를 7로 수정하세요
mydict['green'] = 7
print(mydict['green'])

#mydict dictionary를 출력하세요
print(mydict)

#kye 값에 blue가 포함되어 있는지 확인하세요
'blue' in mydict.keys() 

#mydic의 길이를 출력하세요
print(len(mydict))

#복사해서 mydict2를 만드세요
mydict2 = mydict.copy()
print(mydict2)

#get을 이용해서 red의 value값을 출력하세요
print(mydict.get('red'))

#kye값만 리스트로 추력하세요
print(mydict.keys())

#value값들만 리스트로 출력하세요
print(mydict.values())

#key와 value의 튜플 쌍으로 각 요소를 출력하세요
print(mydict.items())

#key ;black;의 value를 출력하고 없으면 key black와 value 5로 삽입하세요
print(mydict.setdefault('black',5))

#key를 기준으로 정렬하세요
list_key = sorted(mydict.items())
print(list_key)

#value를 기준으로 정렬하세요
list_val=sorted(mydict.items(),key=lambda x:x[1])
print(list_val)

#키 blue와 value를 삭제하세요
del mydict['blue']
print(mydict)

#red를 삭제하면서 value값을 동시에 리턴하세요
print(mydict.pop('red'))

#모든 요소를 삭제하세요
mydict.clear()
print(mydict)

#mydict를 삭제하세요
del mydict
print(mydict)
```
