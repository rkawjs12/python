```python
solarsys = ['태양', '수성', '금성', '지구', '화성', '목성', '토성', '천왕성', '해왕성']
sol_list = []
for i in range(len(solarsys)):
    k = (i, solarsys[i])
    sol_list.append(k)
print(sol_list)
```
